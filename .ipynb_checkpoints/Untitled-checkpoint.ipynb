```python
import pandas as pd

savant2018 = pd.read_csv('D://statcast//savant2018.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
savant2018.head()
```

```python
sam = savant2018[savant2018.player_name == 'Jeurys Familia']
```

```python
sam.head()
```

```python
sam.columns
```

```python
samcol = sam[['pitch_type', 'release_speed', 'release_spin_rate', 'release_extension']]
samcol.head()
```

```python
samcol.groupby('pitch_type').mean()
```

```python
plate = sam[['plate_x', 'plate_z']]
```

```python
plt.style.use('ggplot')

plt.figure()
plt.subplot(1, 2, 1)
plt.hist(sam['release_extension'])
plt.subplot(1, 2, 2)
plt.plot([-2, -2], [0, 4], color='b')
plt.plot([2, 2], [0, 4], color='b')
plt.plot([-2, 2], [0, 0], color='b')
plt.plot([-2, 2], [4, 4], color='b')

plt.scatter(plate['plate_x'], plate['plate_z'])
sns.jointplot(x='plate_x', y='plate_z', data=plate, kind='kde')
```
