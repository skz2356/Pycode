```python
import pandas as pd
```

```python
data = pd.read_csv('D://statcast//savant2018.csv')
```

```python
data.head()
```

```python
count = data.groupby('player_name').agg({'release_speed':'count'})
```

```python
count.head()
```

```python
count100 = count[count.release_speed >= 100]
```

```python
count100.head()
```

```python
name = pd.DataFrame(count100.reset_index()['player_name'])
name.head()
```

```python
speed = data.groupby('player_name').agg({'release_speed':'mean'})
speed_mean = speed.reset_index()
```

```python
speed.info()
```

```python
speed_mean.head()
```

```python
speed_mean_name = speed_mean.merge(name, on='player_name', how='inner')
```

```python
speed_mean_name.sort_values(by='release_speed')
```
